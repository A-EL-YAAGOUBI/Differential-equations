## This tutorial aims to understand the following simple looking equation : 
$
\begin{cases}
    \dot y = cos(\pi xy)\\
    y : x \to y(x)
\end{cases} \hspace{1cm} (E)
$

Although equation $(E)$ looks simple it is a very difficult one. The reason why this equation is difficult is because of the non-linearity caused by the Cosine function, there is no simple way to write down the solution to this equation. First we are going to visualize the solution numericaly and then we are going to attemp to understand we have visualized using hyper asymptotics a.k.a. asymptotics beyond all orders ... 
- I  - Numerical solution
- II - Hper Asymptotics 

```python
import sys
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
```

# I  - Numerical solution

```python
nx, ny = 100, 100            # size of discretization grid
Xl, Xr = 0, 4.5              # location of the grid (x-axis) left, right
Yd, Yu = 0, 2.5              # location of the grid (y-axis) down, up

# building the grid
x = np.linspace(Xl, Xr, nx)
y = np.linspace(Yd, Yu, ny)
X, Y = np.meshgrid(x, y)     # grid

# building the velocity field
Vx = np.ones((ny, nx))
Vy = np.cos(np.pi * X * Y)   # Y dot (slope of y at x)
```

##### Visualization of the integral curves of equation (E)

```python
fig = plt.figure(figsize=(10, 15))
ax = fig.add_subplot(111)
color = 2 * np.log(np.hypot(Vx, Vy))

# Behind the Streamplot function there is a Runge-Kutta 
# scheme and linear interpolation methods are used here 
# to find the integral curves of the velocity field ...
ax.streamplot(x, y, Vx, Vy, color=color, linewidth=1.5, cmap=plt.cm.winter, density=1, arrowstyle='->', arrowsize=1.5)

ax.set_xlabel('$X-Axis$')
ax.set_ylabel('$Y-Axis$')
ax.set_xlim(Xl, Xr)
ax.set_ylim(Yd, Yu)
ax.set_aspect('equal')
plt.title('Integral curves of the trivial looking equation (E)')
plt.show()
```
